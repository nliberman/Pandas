```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("../data/titanic_train.csv")
df.head()
```

```python
df.dtypes
```

## Plot Scatter Matrix on Numeric Columns

```python
pd.scatter_matrix(df)
```

## Plot Scatter Matrix on Specific Columns

```python
pd.scatter_matrix(df[['Age','Fare', 'Survived']], figsize=(12,10))
```

## Plot Value Counts of a Class

```python
df.Pclass.value_counts().plot(kind='bar')
```

## Plot Value Counts of a Class (With Filters)

```python
df[df.Survived==1].Pclass.value_counts().plot(kind='bar')
```
