```python
import pandas as pd
```

```python
df = pd.read_csv("../data/titanic_test.csv")
df.head()
```

```python
df.dtypes
```

## Simple Numerical Filter

```python
df[df.Age>50].head()
```

## Full String Match Filter

```python
df[df.Embarked=="Q"].head()
```

## Numeric and String Filter

```python
df[(df.Embarked=="Q") & (df.Age<=20)].head()
```

```python
df[(df.Embarked=="Q") | (df.Age<=20)].head()
```

## Partial String Match

```python
df[df.Ticket.str.contains("C.A. 3")].head()
```

```python
df[df.Ticket.str.contains("C.A\. 3")].head()
```

## Filter If Matches a List Element

```python
df[df.SibSp.isin([0,2])].head()
```

## Compound Query

```python
df[(df.Embarked=="Q") & (df.Age<=20) & (df.SibSp.isin([0,1])) & df.Ticket.str.contains("Q/3")]
```

## Filter and Return Separate Row

```python
df[(df.Embarked=="Q") | (df.Age<=20)][['Fare']].head()
```

## Aggregate Mean on Filter

```python
df[(df.Embarked=="Q") | (df.Age<=20)]['Fare'].mean()
```

## Retrieve a Value on a filter

```python
df[df.Ticket=='SC/PARIS 2167'].values
```

```python
df[df.Ticket=='SC/PARIS 2167'].values[0][4]
```

## Return Cell on a Filter Which can Only Yield One Row

```python
df[df.Ticket=='SC/PARIS 2167'].reset_index()
```

```python
df[df.Ticket=='SC/PARIS 2167'].reset_index().loc[0,"Age"]
```
