```python
import pandas as pd
```

```python
df = pd.read_csv("../data/titanic_test.csv")
```

```python
df.head()
```

```python
df['Fare'] = df.Fare.apply(lambda x: x+1)
df.head()
```

```python
df['Fare'] = df.Fare.apply(lambda x: x*2-1)
df.head()
```

```python
df['last_name'] = df.Name.apply(lambda x: x.split(", ")[0])
```

```python
df.head()
```

```python
def edit_name(x):
    x = x.split(", ")[0].upper()
    x = "-" + x + "-"
    return x
```

```python
df['last_name_upper'] = df.Name.apply(lambda x: edit_name(x))
```
