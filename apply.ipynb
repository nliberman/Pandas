```python
import pandas as pd
```

```python
df = pd.read_csv("../data/titanic_test.csv")
df.head()
```

## Edit Existing Column

```python
df['Fare'] = df.Fare.apply(lambda x: x+1)
df.head()
```

```python
df['Fare'] = df.Fare.apply(lambda x: x*2-1)
df.head()
```

## Create New Column In One line

```python
df['last_name'] = df.Name.apply(lambda x: x.split(", ")[0])
df.head()
```

## Create New Column with Help of Function

```python
def edit_name(x):
    x = x.split(", ")[0].upper()
    x = "-" + x + "-"
    return x
```

```python
df['last_name_upper'] = df.Name.apply(lambda x: edit_name(x))
df.head()
```
