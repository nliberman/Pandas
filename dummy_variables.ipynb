## Manual Dummy

```python
def q_dummy(x):
    if x == "Q":
        return 1
    else:
        return 0
```

```python
df['Embarked_Q'] = df.Embarked.apply(lambda x: q_dummy(x))
df.head()
```

## With Pandas

```python
df = pd.read_csv("../data/titanic_test.csv")
df.head()
```

```python
pd.get_dummies(df.Embarked).head()
```

```python
dummies = pd.get_dummies(df.Embarked, drop_first=True, prefix="Embarked", prefix_sep="_").head()
dummies.head()
```

## Add Dummy Columns Inplace

```python
df = pd.concat([df, pd.get_dummies(df.Embarked, drop_first=True, prefix="Embarked", prefix_sep="_")], axis='columns')
df.head()
```

## Drop Original Column

```python
df = df.drop(['Embarked'], axis='columns')
df.head()
```
